```python
from data import Point_Generator
from line import Line
from ransac import RANSAC
import matplotlib.pyplot as plt

point_gen = Point_Generator(100, 0.3)
points = point_gen.generate_case(1, 2, eps=0.1)
line = Line(points)
line.estimate_params()
inlnrs, outlnrs = line.devide_points(points, eps=0.1)
k, b = line.get_params()
plt.plot(inlnrs.T[0], inlnrs.T[1], "x", label="inliers")
plt.plot(outlnrs.T[0], outlnrs.T[1], "x", label="outliers")
plt.plot(points.T[0], k * points.T[0] + b, "r", label="Fitted line")
plt.legend()
print("Без RANSAC")
plt.show()
```
