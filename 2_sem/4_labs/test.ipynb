```python
import data
import line
import matplotlib.pyplot as plt
```

```python
point_generator = data.Point_Generator(100, 0.2)
x, y = point_generator.generate_case(None,None)
```

```python
Line = line.Line(x,y)
Line.estimate_params()
inliers_x, inliers_y, outliers_x, outliers_y = Line.devide_points(x, y, eps = 0.5)
k, b = Line.get_params()
```

```python
plt.plot(inliers_x, inliers_y, 'o', label='inliers')
plt.plot(outliers_x, outliers_y, 'o', label='outliers')
plt.plot(x, k*x + b, 'r', label='Fitted line')
plt.legend()
plt.show()
```
